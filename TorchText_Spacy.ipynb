```python
# Reference 
# https://towardsdatascience.com/how-to-use-torchtext-for-neural-machine-translation-plus-hack-to-make-it-5x-faster-77f3884d95

import spacy
import torchtext
```

```python
# en is something like a function that takes a string as input and outputs tokens
en = spacy.load('en_core_web_sm')
de = spacy.load('de_core_news_sm')

# Tokenizer example:
# Sentence: 'This is torchtext tutorial that uses spacy'
# Tokens:  ['This', 'is', 'torchtext', 'tutorial', 'that', 'uses', 'spacy']

# Example
text = en("spaCy is an open-source software library for advanced natural language processing")
print([t for t in text])

# tokens have attributes that help us to extract its properties
print(text.lang_)
```

```python
for t in en("This is a torchtext tutorial that uses spacy!"):
    print(t, t.pos_, t.dep_)
```

```python
tmp = en('This is a torchtext tutorial that uses spacy')
tmp_de = de('Vielen Dank für Ihre Mithilfe und die besten Wünsche für die Feiertage')
spacy.displacy.render(tmp)
spacy.displacy.render(tmp_de)
```
